```python
d_type = []
unqs = []
for i in obsv.columns:
    # print(i)
    # print(len(obsv[i].unique()))
    # print(' ')
    d_type.append(i)
    unqs.append(len(obsv[i].unique()))
    # break
```

```python
unqs
```

```python
d_type
```

```python
d_type_counts = pd.DataFrame(dict(Data_type=d_type,Unique_value_count=unqs)).sort_values('Unique_value_count', ascending  = False)
```

```python
d_type_counts
```

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib.pyplot import figure
```

```python
len(d_type_counts)
```

```python
list(range(0,len(d_type_counts)))
```

```python
import matplotlib
```

```python
from matplotlib import rcParams
```

```python
figure(figsize=(12, 4), dpi=150)
# ax = plt.figure()
plt.bar(d_type_counts.Data_type,
        d_type_counts.Unique_value_count,
       log = True
       )
plt.title('ACED Synthetic Data, Unique Values by Type',
             fontsize = 12)
plt.xticks(list(range(0,len(d_type_counts))), labels = d_type_counts.Data_type, rotation = -45,
           fontsize = 6., ha='left',
           rotation_mode="anchor",
           # va = 'bottom'
           # labelpad = 1
          )
# ax.tick_params(axis='x', which='major', pad=15)
# matplotlib.axes.Axes.tick_params(self, 
#                                  axis = 'x', pad = 5)

rcParams['xtick.major.pad']='-1'
plt.ylabel('Unique values, log scale', fontsize = 11)
plt.xlabel('Data Type', fontsize = 11)
plt.savefig('unique_values_by_datatype.png', dpi=180, bbox_inches = 'tight')
```
